```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://internshala.com/fresher-jobs"
page1 = requests.get(url)
page1
```

```python
soup = BeautifulSoup(page1.content)
soup
```

```python
print(soup.prettify)
```

```python
job_title = [] #Empty list
company_name = [] #Empty list
ctc = [] #Empty list
apply_date = [] #Empty list
```

```python
job_title = []
jb = soup.find_all("div", class_ = "heading_4_5 profile")
for i in jb:
    job_title.append(i.text.replace("\n", ""))
job_title
```

```python
company_name = []
cmp = soup.find_all("a", class_ = "link_display_like_text")
for i in cmp:
    company_name.append(i.text.replace("\n", ""))
company_name
```

```python
CTC = []
ctc = []
ct = soup.find_all("div", class_ = "item_body")
for i in ct:
    ctc.append(i.text.replace("\n", ""))
    
for i in range(1,len(ctc),3):
    CTC.append(ctc[i])
CTC
```

```python
apply_date = []
Apply_date = []
ad = soup.find_all("div", class_ = "item_body")
for i in ad:
    apply_date.append(i.text)

for i in range(2,len(apply_date),3):
    Apply_date.append(apply_date[i])
Apply_date
```

```python
print(len(Apply_date),len(CTC),len(company_name),len(job_title))
```

```python
import pandas as pd
df = pd.DataFrame({})
df["job_title"] = job_title
df["Company_name"] = company_name
df["CTC"] = CTC
df["Apply_date"] = Apply_date

df
```
